```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-2, 2, 100)
y = x ** 2
z = np.sqrt(4 - x**2)
plt.figure(figsize=(4, 4))
plt.plot(x, y, label='$y=x^2$', color='red', linewidth=2)
plt.plot(x, z, 'b--', label='$x^2+y^2=4$')
plt.plot(x, -z, 'y--', label='$x^2+y^2=4$')
plt.xlabel('x轴', font='SimHei')
plt.ylabel('y轴', font='SimHei')
plt.title('线性图', font='SimHei')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid()
plt.savefig('ch_plot.png', format='png', dpi=500)
plt.show()
```
