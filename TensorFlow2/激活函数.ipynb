```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.axisartist as axisartist

def threshold():
    fig = plt.figure(figsize=(6, 6))
    ax = axisartist.Subplot(fig, 111)
    fig.add_axes(ax)
    ax.axis[:].set_visible(False)
    ax.axis['x'] = ax.new_floating_axis(0, 0)
    ax.axis['y'] = ax.new_floating_axis(1, 0)
    ax.axis['x'].set_axisline_style('-|>', size=1.0)
    ax.axis['y'].set_axisline_style('-|>', size=1.0)
    ax.axis['x'].set_axis_direction('top')
    ax.axis['y'].set_axis_direction('right')
    plt.ylim(-0.2, 1.25)
    x_1 = np.arange(0, 10, 0.1)
    y_1 = x_1 - x_1 + 1
    plt.plot(x_1, y_1, 'r', label=r'threshold')
    plt.legend()
    x_2 = np.arange(-5, 0, 0.1)
    y_2 = x_2 - x_2
    plt.plot(x_2, y_2, 'r', label='threshold')
    plt.scatter(0, 1, color='r')
    plt.scatter(0, 0, marker='o', color='', edgecolors='r')
    plt.savefig('./threshold.png', format='png')
    
threshold()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.axisartist as axisartist

def sigmoid():
    fig = plt.figure(figsize=(6, 6))
    ax = axisartist.Subplot(fig, 111)
    fig.add_axes(ax)
    ax.axis[:].set_visible(False)
    ax.axis['x'] = ax.new_floating_axis(0, 0)
    ax.axis['y'] = ax.new_floating_axis(1, 0)
    ax.axis['x'].set_axisline_style('-|>', size=1.0)
    ax.axis['y'].set_axisline_style('-|>', size=1.0)
    ax.axis['x'].set_axis_direction('top')
    ax.axis['y'].set_axis_direction('right')
    plt.xlim(-10, 10)
    plt.ylim(-0.1, 1.2)
    x = np.arange(-10, 10, 0.1)
    y = 1 / (1+np.exp(-x))
    plt.plot(x, y, label=r'sigmoid', c='r')
    plt.legend()
    plt.savefig('sigmoid.png', format='png')
    
sigmoid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.axisartist as axisartist

def relu():
    fig = plt.figure(figsize=(6, 6))
    ax = axisartist.Subplot(fig, 111)
    fig.add_axes(ax)
    ax.axis[:].set_visible(False)
    ax.axis['x'] = ax.new_floating_axis(0, 0)
    ax.axis['y'] = ax.new_floating_axis(1, 0)
    ax.axis['x'].set_axisline_style('-|>', size=1.0)
    ax.axis['y'].set_axisline_style('-|>', size=1.0)
    ax.axis['x'].set_axis_direction('top')
    ax.axis['y'].set_axis_direction('right')
    plt.xlim(-10, 10)
    plt.ylim(-0.1, 10)
    x_1 = np.arange(0, 10, 0.1)
    y_1 = x_1
    plt.plot(x_1, y_1, 'r-', label=r'relu')
    x_2 = np.arange(-5, 0, 0.1)
    y_2 = x_2 - x_2
    plt.plot(x_2, y_2, 'r-')
    plt.legend()
    plt.savefig('relu.png', format='png')
    
relu()
```
