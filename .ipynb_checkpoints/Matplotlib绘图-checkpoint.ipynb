# 绘制线性图

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-2, 2, 100)
y = x ** 2
z = np.sqrt(4 - x**2)
plt.figure(figsize=(4, 4))
plt.plot(x, y, label='$y=x^2$', color='red', linewidth=2)
plt.plot(x, z, 'b--', label='$x^2+y^2=4$')
plt.plot(x, -z, 'y--', label='$x^2+y^2=4$')
plt.xlabel('x轴', font='SimHei')
plt.ylabel('y轴', font='SimHei')
plt.title('线性图', font='SimHei')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid()
plt.savefig('out.png', format='png', dpi=500)
plt.show()
```

# 坐标轴带箭头图像

```python
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist
import numpy as np
fig = plt.figure(figsize=(6, 6))
ax = axisartist.Subplot(fig, 111)
fig.add_axes(ax)
ax.axis[:].set_visible(False)
ax.axis['x'] = ax.new_floating_axis(0, 0)
ax.axis['y'] = ax.new_floating_axis(1, 0)
ax.axis['x'].set_axisline_style('- | >', size=1.0)
ax.axis['y'].set_axisline_style('- | >', size=1.0)
ax.axis['x'].set_axis_direction('top')
ax.axis['y'].set_axis_direction('right')
plt.xlim(-10, 10)
plt.ylim(-0.1, 1.2)
x = np.arange(-10, 10, 0.1)
y = 1 / (1+np.exp(-x))
plt.plot(x, y, label=r'$sigmiod=\frac{1}{1+e^{-x}}$', c='r')
plt.legend()
plt.savefig('sigmoid.png', format='png', dpi=500)
```
