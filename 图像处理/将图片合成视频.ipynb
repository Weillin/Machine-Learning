```python
import os
import glob as gb
import cv2


if not os.path.exists('out'):
    os.mkdir('out')
vidcap = cv2.VideoCapture('3.mp4')

success, image = vidcap.read()
fps = int(vidcap.get(cv2.CAP_PROP_FPS))

count = 0
while success:
    if fps != 0:
        print(count)
        cv2.imwrite('out/%d.jpg' % int(count), image)

    success, image = vidcap.read()
    count += 1

path = gb.glob('out\*.jpg')
img_path = []
img_shape = (1280, 720)
for i in range(len(path)):
    img_path.append('out' + '/' + str(i) + '.jpg')
print(img_path)
videoWriter = cv2.VideoWriter('test.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 30, img_shape)

for path in img_path:
    img = cv2.imread(path)
    img = cv2.resize(img, img_shape)
    videoWriter.write(img)
```
