```python
import numpy as np
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
x1 = np.random.uniform(-4, 4,(1,n))
y1 = np.random.uniform(-4, 4,(1,n))
plt.scatter(x1, y1, color='yellow', marker='o')
plt.scatter(x, y,color='blue', marker='*')
plt.legend()
plt.title('标准正态分布', fontsize=20)
plt.text(2.5,2.5, '均值: 0\n标准差: 1')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel('横坐标x', fontsize=14)
plt.ylabel('纵坐标y', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei'
n = 24
y1 = np.random.randint(27, 37, n)
y2 = np.random.randint(40, 60, n)
plt.plot(y1, label='温度')
plt.plot(y2, label='湿度')
plt.xlim(0, 23)
plt.ylim(20, 70)
plt.xlabel('小时', fontsize=12)
plt.ylabel('测量值', fontsize=12)
plt.title('24小时温度湿度统计', fontsize=16)
plt.legend()
plt.show()
```

```python
(train_x, train_y), (_, _) = boston_housing.load_data(test_split=0.2)
plt.figure(figsize=(5,5))
plt.scatter(train_x[:, 5], train_y)
plt.xlabel('RM')
plt.ylabel("Price($1000's)")
plt.title('5.RM-Price')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
for i in range(13):
    plt.subplot(4,4,(i+1))
    plt.scatter(train_x[:,i], train_y)
plt.show()
```

```python
titles = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis','tax','ptratio','b-1000','lstat','medv']
plt.figure(figsize=(12,12))
for i in range(13):
    plt.subplot(4,4,(i+1))
    plt.scatter(train_x[:,i], train_y)
    plt.xlabel(titles[i])
    plt.ylabel("Price($1000's)")
    plt.title(str(i+1)+'.'+titles[i]+'- Price',fontsize=10)
plt.suptitle('各个属性与房价的关系', x=0.5, y=1.02, fontsize=20)
plt.legend()
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
