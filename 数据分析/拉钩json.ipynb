```python
import requests
import jsonpath

def get_json(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'
    }
    res = requests.get(url, headers=headers).json()
    print(res)
    city_list = jsonpath.jsonpath(res, '$..name')
    code_list = jsonpath.jsonpath(res, '$..code')
    id_list = jsonpath.jsonpath(res, '$..id')
    isSelected_list = jsonpath.jsonpath(res, '$..isSelected')
    parentId_list = jsonpath.jsonpath(res, '$..parentId')
    for city, code, id, isSelected, parentId in zip(city_list, code_list, id_list, isSelected_list, parentId_list):
        print(id, city, parentId, code, isSelected)

def main():
    url = 'https://www.lagou.com/lbs/getAllCitySearchLabels.json'
    get_json(url)

if __name__ == '__main__':
    main()
```
