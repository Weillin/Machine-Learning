```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax3d = Axes3D(fig)
plt.show()
```

```python
import numpy as np

x = np.random.uniform(10, 40, 30)
y = np.random.uniform(100, 200, 30)
z = np.random.uniform(10, 20, 30)

fig = plt.figure()
ax3d = Axes3D(fig)
ax3d.scatter(x, y, z, c='b', marker='*')
plt.show()
```

```python
x = np.random.uniform(10, 40, 30)
y = np.random.uniform(100, 200, 30)
z = 2 * x + y

fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.scatter(x, y, z, c='b', marker='*')

ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Z=2X+Y')

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 5, 6]
X, Y = np.meshgrid(x, y)
print(X)
print(Y)
```

```python
x = np.arange(1, 5)
y = np.arange(1, 5)
X, Y = np.meshgrid(x, y)
print(X.shape)
print(Y.shape)
```

```python
Z = 2 * X + Y
print(Z.shape)
```

```python
fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.plot_surface(X, Y, Z, cmap='rainbow')

ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Z=2X+Y')
plt.show()
```

```python
x = np.arange(0, 10)
y = np.arange(0, 10)
X, Y = np.meshgrid(x, y)
Z = 2 * X + Y

fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.plot_surface(X, Y, Z, cmap='rainbow')

ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Z=2*X+Y')
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.arange(0, 10, 0.1)
X, Y = np.meshgrid(x, y)
Z = 2 * X + Y
print(Z.shape)
```

```python
fig = plt.figure()
ax3d = Axes3D(fig)

surf = ax3d.plot_surface(X, Y, Z, cmap='rainbow')
fig.colorbar(surf, shrink=0.5, aspect=5)

ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Z=2X+Y')
plt.show()
```

```python
x = np.arange(1, 10, 0.5)
y = np.arange(1, 10, 0.5)
X, Y = np.meshgrid(x, y)
Z = 2 * X + Y

fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.plot_wireframe(X, Y, Z, color='m', linewidth=0.5)

ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Z=2X+y')
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))

fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.plot_surface(X, Y, Z, cmap='rainbow')
plt.show()
```

```python
x1 = np.array([137, 104, 100, 124, 79, 99, 124, 114.0])
x2 = np.array([3, 2, 2, 3, 1, 2, 3, 2])
y = np.array([145, 110, 93, 116, 65, 104, 118, 91.0])

W = np.array([11.995, 0.5352, 14.35793])
y_pred = W[1] * x1 + W[2] * x2 + W[0]

fig = plt.figure(figsize=(8, 6))
ax3d = Axes3D(fig)

ax3d.scatter(x1, x2, y, color='b', marker='*')

ax3d.set_xlabel('Area', color='r', fontsize=16)
ax3d.set_ylabel('Room', color='r', fontsize=16)
ax3d.set_zlabel('Price', color='r', fontsize=16)
ax3d.set_yticks([1, 2, 3])
ax3d.set_zlim3d(30, 160)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax3d = Axes3D(fig)
ax3d.view_init(elev=0, azim=-90)

ax3d.scatter(x1, x2, y, color='b')

ax3d.set_xlabel('Area', color='r', fontsize=16)
ax3d.set_ylabel('Room', color='r', fontsize=16)
ax3d.set_zlabel('Price', color='r', fontsize=16)
ax3d.set_yticks([1, 2, 3])
ax3d.set_zlim3d(30, 160)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax3d = Axes3D(fig)
ax3d.view_init(elev=0, azim=0)

ax3d.scatter(x1, x2, y, color='b')

ax3d.set_xlabel('Area', color='r', fontsize=16)
ax3d.set_ylabel('Room', color='r', fontsize=16)
ax3d.set_zlabel('Price', color='r', fontsize=16)
ax3d.set_yticks([1, 2, 3])
ax3d.set_zlim3d(30, 160)

plt.show()
```

```python
X1, X2 = np.meshgrid(x1, x2)
Y_pred = W[0] + W[1] * X1 + W[2] * X2

fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.plot_surface(X1, X2, Y_pred, cmap='coolwarm')

ax3d.set_xlabel('Area', color='r', fontsize=14)
ax3d.set_ylabel('Room', color='r', fontsize=14)
ax3d.set_zlabel('Price', color='r', fontsize=14)
ax3d.set_yticks([1, 2, 3])

plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

fig = plt.figure()
ax3d = Axes3D(fig)

ax3d.scatter(x1, x2, y, color='b', marker='*', label='')
ax3d.scatter(x1, x2, y_pred, color='r', label='')
ax3d.plot_wireframe(X1, X2, Y_pred, color='c', linewidth=0.5, label='')

ax3d.set_xlabel('Area', color='r', fontsize=14)
ax3d.set_ylabel('Room', color='r', fontsize=14)
ax3d.set_zlabel('Price', color='r', fontsize=14)

plt.suptitle('', fontsize=20)
plt.legend(loc='upper left')
plt.show()
```
