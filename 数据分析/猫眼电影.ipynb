```python
import requests
import json
from lxml import etree
import time

def getOnePage(n):
    url = f'https://maoyan.com/board/4?offset={n*10}'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36'}
    time.sleep(5)
    r = requests.get(url,headers=headers)
    return r.text

def parse(text):
    html = etree.HTML(text)
    names = html.xpath('//div[@class="movie-item-info"]/p[@class="name"]/a/@title')
    releasetime = html.xpath('//p[@class="releasetime"]/text()')
    item = {}
    for name, rele in zip(names, releasetime):
        item['name'] = name
        item['releasetime'] = rele
        yield item

def saveFile(data):
    with open('movie.json', 'a', encoding='utf-8') as f:
        data = json.dumps(data, ensure_ascii=False) + ',\n'
        f.write(data)

def run():
    for n in range(0, 10):
        text = getOnePage(3)
        items = parse(text)
        for item in items:
            print(item)
            saveFile(item)

if __name__ == '__main__':
    run()
```
