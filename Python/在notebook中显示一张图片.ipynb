## 示例一：使用opencv打开

### 1.导入依赖包

```python
from matplotlib import pyplot as plt
import cv2
```

### 2.设置matplotlib在线显示

```python
%matplotlib inline
```

### 3.用opencv读取图片

```python
img= cv2.imread('../../data/input.png')
print(img)
```

### 4.opencv读取图片是BGR，matplotlib读取图片是RGB，所以需要将矩阵颠倒过来

```python
img2 = img[:,:,::-1]
```

### 5，显示图片

```python
plt.imshow(img2)
```
